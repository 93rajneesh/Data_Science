#### Read updated dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

#### Task 2: Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2] # month in str data type
all_data.head()
```

### Cleaning
#### clean Nan values

```python
 nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()
```

```python
all_data = all_data.dropna(axis =0, how = 'all' )
all_data
```

#### Find 'or' and delete it

```python
all_data = all_data[all_data['Month'].str[0:2] !='Or']
all_data
```

### Task 3 Add sales column

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # float

all_data['Sales'] = all_data['Quantity Ordered']* all_data['Price Each']
all_data.head()
```

### Add city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + " " + get_state(x))
all_data
```

### Q2 Which city has highest number of sales?

```python
sale_city = all_data.groupby('City').sum()
sale_city
```

```python
cities = [city for city , df in all_data.groupby('City')]
plt.bar(cities , sale_city['Sales'])
plt.xlabel("City")
plt.ylabel("Sales in Million USD")
plt.xticks(cities , rotation = 'vertical' , size = 8)
plt.show()
```

### Q3: What time should we display advertisment to maximize likehood of customer's buying products? 

```python
def hour(time):
    x = time.split(' ')[1]
    return x.split(':')[0]
def minute(time):
    x = time.split(' ')[1]
    return x.split(':')[1]


all_data['Hour'] = all_data['Order Date'].apply(lambda x : hour(x))
all_data['Minute'] = all_data['Order Date'].apply(lambda x : minute(x))
```

```python
all_data.head()
```

```python
hours = [hour for hour , d in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xlabel("Hour")
plt.ylabel("Count")
plt.grid()
plt.xticks(hours , rotation ='horizontal' , size = 10 )
plt.show()
```
