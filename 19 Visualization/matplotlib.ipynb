```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 , 200)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
# labelling x and y
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is title")
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
plt.scatter(x1,y1)
```

```python
x1
```

```python
y1
```

```python
# colouring 
# size of x,y , colours and size should be same
colours = np.random.rand(50)
size = 500*np.random.rand(50)
plt.scatter(x1,y1 , c = colours , s = size, alpha = .5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
# coordinate represent by differnt colours
x7 = [3,5,8,0,2,3,5,7]
y7= [1,2,4,4,5,6,7,9]
colour = ['red' , 'black', 'pink' , 'purple' , 'orange' ,'skyblue' , 'green' ,'grey']
plt.scatter(x7,y7 , c = colour)
plt.show()
```

```python
# 3d projection
x7 = np.random.rand(50)
y7 = np.random.rand(50)
z7 = np.random.rand(50)
#figure(): Create a new figure, or activate an existing figure
fig = plt.figure()
axis = fig.add_subplot(projection = '3d')
axis.scatter(x7,y7,z7 ,c = 'red' , marker = 'p')
plt.show()

```
