**Q1. How can you create a Bokeh plot using Python code?**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

# create output_file
output_file("bokeh.html")
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
p = figure(title="bokeh assignment")
p.line(x, y)
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?



```python
from bokeh.models.plots import Legend
x = [1,2,3,4,5]
y = [6,7,8,9,10]
output_file("Q3.html")
p = figure( title = 'customize' ,  sizing_mode ='stretch_width' )
p.circle(x,y , legend_label = 'red points' , size = 20 , color = 'purple')
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Ans: **The primary purpose** of Bokeh server to synchronize data between the underlying python environment and the BokehJS library in the browser.

Bokeh server makes it eay to create interactive web application that connect front-endUI events to running python code.

Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS.





Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Ans: **server_document() function:** This function accepts the URL to a Bokeh server application and returns a script that embeds a new session from that server every time the script executes. You can add this tag to an HTML page to include the Bokeh application at that point.
