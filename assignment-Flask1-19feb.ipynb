### Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans. flask is web framework that allows us to build lightweight web applications quickly and easily witth flask libraries. 

#### advantage
1. Flask is lightweight backend framework with minimal dependencies.
2. easy to learn because its simple and intuitive API makes it easy to learn and use for beginners.
3. flexible framework because it allows you to customize and extend the framework to suit your needs easily.
4. It can be use in any database like- SQL and NoSQL and with any frontend technology such as react and angular.
5. It is great for small and medium project that not require the complexity of large framework.
